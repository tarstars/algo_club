```python
def is_palindrome(s):
    n = len(s)
    for p in range(n // 2):
        if s[p] != s[n - 1 - p]:
            return False
    return True
```

```python
is_palindrome('amanaplanacanalpanama')
```

```python
is_palindrome('abcdef')
```

```python
def is_cool_palindrome(s):
    n = len(s)
    p = 0
    q = n - 1
    while p < q:
        while s[p].isspace() and p < q:
            p += 1
        while s[q].isspace() and p < q:
            q -= 1
        if s[p] != s[q]:
            return False
        p += 1
        q -= 1
    return True
```

```python
is_cool_palindrome('amanap lanac a nalp a nam a')
```

```python
is_cool_palindrome('ambnap lanac a nalp a nam a')
```

```python
def reverse(s):
    s = list(s)
    n = len(s)
    p = 0
    q = n - 1
    while p < q:
        s[p], s[q] = s[q], s[p]
        p += 1
        q -= 1
    return ''.join(s)
```

```python
reverse('amanap lanac a nalp a nam a')
```

```python
def generic_search(v0, f, action):
    todo = [v0]  # stack -> depth-first search, 
    visited = set([v0])
    action(v0)
    
    while todo:
        current = todo.pop()
        action(current)
        for child in f(current):
            if child not in visited:
                visited.add(child)
                todo.append(child)
                
```
