```python
import turtle
import math
```

```python
def norm(p):
    return math.hypot(*p)


def unity(p):
    r = norm(p)
    return tuple(t/r for t in p)


def ccw(p):
    return (-p[1], p[0])


def dot(p1,p2):
    return sum(t1*t2 for t1,t2 in zip(p1,p2))


def mul(p, k):
    return tuple(t*k for t in p)


def plus(p1, p2):
    return tuple(t1+t2 for t1, t2 in zip(p1, p2))


def shift(p):
    turtle.goto(*sum(turtle.position(), p))

    
def line(p1, p2, c=(0, 0, 0)):
    turtle.up()
    turtle.pencolor(c)
    turtle.goto(*p1)
    turtle.down()
    turtle.goto(*p2)
```

```python
turtle.speed(0)
turtle.delay(0.01)
turtle.pensize(2)

for t in range(0, 629*10, 5):
    phi_system = t/100/3
    e11 = (17*math.cos(phi_system), 17*math.sin(phi_system))
    e1 = unity(e11)
    e2 = ccw(e1)
    e1d = mul(e1, 50)
    e2d = mul(e2, 50)
    
    turtle.home()
    turtle.clear()
    phi = t/100
    r=(1, 0, 0)
    g=(0, 1, 0)

    p=(250*math.cos(phi),250*math.sin(phi))
    px=dot(p, e1)
    py=dot(p, e2)
    sh1 = mul(e2, -5)
    sh2 = mul(e1, -5)

    line(plus(sh1, (0, 0)), plus(sh1, e1d), r)
    line(plus(sh2, (0, 0)), plus(sh2, e2d), g)
    
    line((0, 0),  mul(e1,px), r)
    line((0, 0),  mul(e2,py), g)
    line((0,0), p)
    
    turtle.time.sleep(0.1)
```
